```python
import pandas as pd
game_file="Resources/purchase_data.csv"
purchase_data=pd.read_csv(game_file)
purchase_data.head(5)
```

```python
#total player count
player_count = len(purchase_data["SN"].value_counts())
player_count = pd.DataFrame({"Total Players":player_count},index=[0])
player_count
```

```python
#unique items
unique_items= len(purchase_data["Item ID"].value_counts())

#average price of items purchased
average_price= round(purchase_data["Price"].mean(),2)

#number of purchases
purchase_total= purchase_data["Item Name"].count()

#Total Revenue
revenue_total = purchase_data["Price"].sum()

#Create data frame with these calculations

summary_df=pd.DataFrame({"Number of Unique Items":[unique_items],"Average Price":[average_price],
                        "Number of Purchases":[purchase_total], "Total Revenue":[revenue_total]})
summary_df
```

```python
#Group by gender and look for duplicate SN
group_df = purchase_data.groupby(["Gender"])
filtered_df = group_df.nunique()

#Total Players
total_players = filtered_df["SN"].sum()

#Calculate Percentage of Players and count
gender_count = filtered_df["SN"].unique()
gender_pct= round((filtered_df["SN"]/total_players),2).map("{:,.2%}".format)

#Data frame with count and percentage of players grouped by gender

gender_df= pd.DataFrame({"Percentage of Players":gender_pct,
                        "Player Count": gender_count})
gender_df
```

```python
#Spending habits by gender
purchase_count = purchase_data["Gender"].value_counts()

#avg purchase price
average_price = group_df["Price"].mean().map("${:,.2f}".format)

#sum of purchases by gender
sum_price = group_df["Price"].sum()

#Average purchase total per person by gender
avg_gender = round(sum_price/gender_count,2).map("${:,.2f}".format)

gender_spending = pd.DataFrame({"Purchase Count":purchase_count,
                                "Average Purchase Price":average_price,
                                "Total Purchases":sum_price,
                               "Average Spending by Gender":avg_gender})
gender_spending
```

```python
#Age Analysis
#Create Bins and Names for them
bins = [0,10,14,18,22,26,30,34,38,150]
bin_names=["<10","11-14","15-18","19-22","23-26","27-30","31-34","35-38","39+"]

purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels = bin_names)
age_df = purchase_data.groupby(["Age Groups"])

#Purchase count
purchase_count = purchase_data["Age Groups"].value_counts()

#Total Purchase Value
total_purchase = round(age_df["Price"].sum(),2)

#Average Purchase Price
avg_purchase = age_df["Price"].mean()


#Average Purchase Total per Person by Age Group
avg_price_age = round(total_purchase / purchase_count,2).map("${:,.2f}".format)

#Create final data frame with age data
final_age_df = pd.DataFrame({"Purchase Count":purchase_count,
                             "Average Purchase Price":avg_purchase.map("${:,.2f}".format),
                             "Total Purchase Value":total_purchase.map("${:,.2f}".format),
                             "Average Purchase per Person":avg_price_age})
final_age_df
```

```python
#Top Spenders
grouped_df = purchase_data.groupby(["SN"])

#Total Purchase Value
total_spent_user = grouped_df.sum()["Price"]

#Purchase Count per user
purchase_count_user = grouped_df["SN"].count()

#Average Purchase Price
avg_user_purchase = (total_spent_user/purchase_count_user).map("${:,.2f}".format)

#Create data frame with these values
spenders_df = pd.DataFrame({"Purchase Count":purchase_count_user,
                            "Average Purchase Price":avg_user_purchase,
                            "Total Purchase Value":total_spent_user})
#Sort by biggest spenders
user_sorted = spenders_df.sort_values("Total Purchase Value", ascending = False)
user_sorted.head(5)


```

```python
#Most popular items analysis
most_items = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
most_items.columns = ['Purchase Count', 'Total Purchase Value']

#Sort by most popular items in Purchase Count
most_popular_items = most_items.sort_values(by='Purchase Count', ascending=False)
most_popular_items.head()
```

```python
#in descending order 
most_profitable = most_items.sort_values(by='Total Purchase Value', ascending=False)
most_profitable.head()
```
